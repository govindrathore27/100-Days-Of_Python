```python
from random import randint 
from art_Number_Scholar import logo
EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5
SPHYNX = """
-=[ Sphinx ]=-  5/98


                              .sSSSSSSSs
                              sSS=""^^^"s
                  /\       , /  \_\_\|_/_)
                 /';;     /| \\\/.-. .-./
                / \;|    /. \,S'  -   - |
               / -.;|    | '.SS     _|  ;
              ; '-.;\,   |'-.SS\   __  /S
              | _  ';\\.  \' SSS\_____/SS
              |  '- ';\\.  \_SSS[_____]SS
              \ '--.-';;-. __SSS/\    SSS
               \  .--' ';;'.=SSS`\\_\_SSS
                `._ .-'` _';;..=.=.=.\.=\
                   ;-._-'  _.;\.=.=.=.|.=|
         ,     _.-'    `"=._  ;\=.=__/__/
         )\ .'`   __        ".;|.=.=.=./
         (_\   .-`  '.   |    \/=.=.=/`
          /\\         \-,|     |.--'|
         /  \`,       //  \    | |  |
        ( (__) )  _.-'--,  \   | |  '--,
         ;----' -'--,__}}}  \  '--, __}}}
  jgs    \_________}}}       \___}}}



"""
def check_answer(guess,answer,turns):
    '''
    Function to check whether answer is right or not.If wrong then turns is reduced by 1.
    '''
    if guess > answer:
        print("Too High")
        return turns - 1
    elif guess < answer:
        print("Too Low")
        return turns - 1
    else:
        print(f"Wow you guessed correctly,you won , answer is {answer}")
def set_difficulty():
    '''
    Function to set difficulty of the game.
    '''
    level = input("Enter the difficulty level you want to play.Easy or Hard.")
    if level.lower() == 'easy':
        return EASY_LEVEL_TURNS
    else:
        return HARD_LEVEL_TURNS
def game():
    print(SPHYNX)
    print("\n ")
    print(logo)
    # Choosing number between 1 and 100
    print("Welcome to the number scholar.")
    print("I will be choosing a number between 1 and 100.You have to the guess the number.\n Guess wisely.")
    answer = randint(1,100)
    guess = 0
    turns = set_difficulty()
    while guess!=answer:
        # taking input for the guess
        guess = int(input("Enter you choice a number between 1 and 100:\n"))
        #Track the number of turns and reduce by 1 if they get it wrong.
        turns = check_answer(guess, answer, turns)
        if turns == 0:
            print("You've run out of guesses, you lose.")
            return
        elif guess != answer:
            print("Guess again.")

game()
print("See you next time!!")
    

    
```
